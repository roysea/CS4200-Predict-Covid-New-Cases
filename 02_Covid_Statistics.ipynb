```python
import seaborn as sns
sns.set_theme()
```

```python
# Plot new cases in 52 states
df.groupby('state').mean().plot.bar(y='newCases',
            rot=45,
            figsize=(15,5), # Figsize to make the plot larger
            title='Number of New Cases per State', # Adding a title to the top
            xlabel="States", # Adding a label on the xaxis
            ylabel="New Cases", # Adding a label on y axis
            fontsize='small'
            ); 
```

```python
# Plot new cases in 9 US Divisions
bar = df.groupby('division').mean().plot.bar(y='newCases',
            rot=0,
            figsize=(15,5), # Figsize to make the plot larger
            title='Number of New Cases per Division', # Adding a title to the top
            xlabel="Division", # Adding a label on the xaxis
            ylabel="New Cases", # Adding a label on y axis
            fontsize='small'
            ); 
bar.set_xticklabels( ('New England', 'Middle Atlantic','East North','West North Central',
                      'South Atlantic','East South Central','West South Central','Mountain', 'Pacific') )
```

```python
# Plot new cases in 4 US Regions
bar = df.groupby('region').mean().plot.bar(y='newCases',
            rot=0,
            figsize=(8,5), # Figsize to make the plot larger
            title='Number of New Cases per US Region', # Adding a title to the top
            xlabel="Region", # Adding a label on the xaxis
            ylabel="New Cases", # Adding a label on y axis
            fontsize='small'
            ); 
bar.set_xticklabels( ('Northeast', 'Midwest','South','West') )
```

```python
# Plot new cases by presence of statewide mask mandate and without mask mandate

bar = df.groupby('mask_mandate').mean().plot.bar(y='newCases',
            rot=0,
            figsize=(8,5), # Figsize to make the plot larger
            title='Number of New Cases with and without state mask mandate', # Adding a title to the top
            xlabel="Mask Mandate", # Adding a label on the xaxis
            ylabel="New Cases", # Adding a label on y axis
            fontsize='small'
            ); 
bar.set_xticklabels( ('Without Mask Mandate', 'With Mask Mandate') )
```

Number of New Cases with and without statewide mask mandate

```python
result = df.groupby(['state', 'mask_mandate'])['newCases'].mean().unstack().plot.bar(rot=45,
            figsize=(15,5), # Figsize to make the plot larger
            title='Number of New Cases with and without statewide mask mandate', # Adding a title to the top
            xlabel="States with and without mask mandate", # Adding a label on the xaxis
            ylabel="New Cases", # Adding a label on y axis
            fontsize=10
)
result.legend(['Average new case count without mask mandate', 'Average new case count with mask mandate'],
              loc=1,fontsize=14)
```

Number of Total Deaths with and without statewide mask mandate

```python
result = df.groupby(['state', 'mask_mandate'])['deaths'].max().unstack().plot.bar(rot=45,
            figsize=(15,5),
            title='Number of Total Deaths with and without statewide mask mandate', # Adding a title to the top
            xlabel="States with and without mask mandate", # Adding a label on the xaxis
            ylabel="Deaths", # Adding a label on y axis
            fontsize='small'
)
result.legend(['Death count without mask mandate', 'Death count with mask mandate'],
              loc=1,fontsize=14)
```
