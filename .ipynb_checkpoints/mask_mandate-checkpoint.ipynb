## Multiple linear regression for number of vaccinated people, location (state of residence) and  number of deaths in that state

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from mpl_toolkits.mplot3d import Axes3D
```

### Data preparation

```python
df.head(10)
```

```python
# change state name to integer
states = df['state'].unique()
print(states)
```

```python
df = df.dropna()
df
```

```python
for i, j in enumerate(states):
    df['state'] = df['state'].replace(j, i)   
df
```

```python
df = df.reset_index()
df
```

```python
X = df[['state','vaccinationsCompleted']].values.reshape(-1,2)
Y = df['deaths'].values
X
```

### Prepare data point for model

```python
x = X[:,0]
y = X[:,1]
z = Y
```

```python
print(y.min(), y.max())
```

### Define and Train Model

```python
x_pred = np.linspace(0, 52, 53)   # states
y_pred = np.linspace(0, 24374216.0, 50)  # range vaccinted people
xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)
model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T
```

```python
ols = linear_model.LinearRegression()
model = ols.fit(X, Y)
predicted = model.predict(model_viz)
```

```python
r2 = model.score(X, Y)
```

```python
plt.style.use('default')

fig = plt.figure(figsize=(12, 10))

ax1 = fig.add_subplot(projection='3d')

ax1.plot(x, y, z, color='k', zorder=15, linestyle='--', marker='o', alpha=0.15)
ax1.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted, facecolor=(0,0,0,0), s=50, edgecolor='#70b3f0')
ax1.set_xlabel('state', fontsize=12)
ax1.set_ylabel('vaccinationsCompleted', fontsize=12)
ax1.set_zlabel('deaths', fontsize=12)
ax1.view_init(elev=28, azim=160)

fig.suptitle('$R^2 = %.2f$' % r2, fontsize=20)
fig.tight_layout()

```

```python
for ii in np.arange(0, 360, 1):
    ax1.view_init(elev=32, azim=ii)
    fig.savefig('images/multiple_regression_vacc_st_death.png')
```
