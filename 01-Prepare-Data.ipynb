```python
# read data from file
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CS4200-Predict-Covid-New-Cases/state_ts.csv")
```

```python
# set vaccination number to 0 for those lines that the result is NaN
df['vaccinationsCompleted'].fillna(0, inplace=True)
```

```python
# remvoe Porto Rico data
# df[df['state']=='PR'].head(10)
df = df.drop(df[df['state']=='PR'].index)
```

```python
#Drop rows that has any NaN data
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
from datetime import datetime

# convert date column to float type
df['date'] = pd.to_datetime(df['date'])

for i in range(len(df['date'])):
  df['date'][i]=df['date'][i].timestamp()

df['date']
```

```python
# get state abbreviations
states = df['state'].unique()
print(len(states))
states
```

```python
# read state region and division from file
st = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CS4200-Predict-Covid-New-Cases/state_region_division.csv")
print(len(st))
st.loc[st['States']=='AK']
```

```python
# insert two columns for region and division
df.insert(1,'region',0,allow_duplicates=True)
df.insert(2,'division',0,allow_duplicates=True)

df.head(3)
```

Add the region and division of each state

```python
for i, k in enumerate(st['States']):
    df.loc[df.state==k,'region'] = st['Region'][i]
    df.loc[df.state==k,'division']= st['Division'][i]
df.head(3)
```

```python
# Test to see if there is any null data in regions
df['region'].isnull().sum()
```
